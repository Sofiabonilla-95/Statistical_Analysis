```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import pdist, squareform
import math
```

```python
movies_df=pd.read_csv('./movie_lens/movies.csv')
```

```python
ratings_df=pd.read_csv('./movie_lens/ratings.csv')
```

```python
ratings_df
```

```python
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))',expand=False)
```

```python
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand=False)
```

```python
movies_df['title'] = movies_df.title.str.replace(' (\(\d\d\d\d\))', '')
```

```python
movies_df.drop(columns=['genres'],inplace=True)
```

```python
movies_df
```

```python
movies_df=movies_df.loc[(movies_df.loc[:,'year']>'1995')]
movies_df
```

```python
userInput = [
{'title':'Titanic', 'rating':5},
{'title':'Mulan', 'rating':5},
{'title':'Crazy, Stupid, Love.', 'rating':4.5},
{'title':'Pirates of the Caribbean: The Curse of the Black Pearl', 'rating':4.5},
{'title':'Disturbia', 'rating':4.5}
]
choice_Movies = pd.DataFrame(userInput)
```

```python
filtro_movies_df=movies_df[np.isin(movies_df, choice_Movies)]
```

```python
choice_movies_1 = pd.merge(filtro_movies_df, choice_Movies)
```

```python
choice_movies_1 
```

```python
choice_movies_1.drop(columns=['year'],inplace=True)
```

```python
choice_movies_1
```

```python
filtro_rating_df=ratings_df[np.isin(ratings_df['movieId'], choice_movies_1['movieId'].tolist())]
filtro_rating_df
```

```python
filtro_rating_df_1 =filtro_rating_df.groupby(['userId'])
```

```python
filtro_rating_df_2 = sorted(filtro_rating_df_1, key=lambda x: len(x[1]), reverse=True)
```

```python
filtro_rating_df_2
```

```python
filtro_rating_df_3 = filtro_rating_df_2[0:100]
```

```python
filtro_rating_df_3
```

```python
pearsonCorrelationDict={}
```

```python
for name, group in filtro_rating_df_3:
    group = group.sort_values(by='movieId')
    choice_movies_1 = choice_movies_1.sort_values(by='movieId')
    movie=choice_movies_1[np.isin(choice_movies_1['movieId'], group['movieId'].tolist())]
    rating = movie['rating'].tolist()
    rating_1 = group['rating'].tolist()
    data_corr = {'tempGroupList': rating_1,
            'tempRatingList': rating}
    df_corr = pd.DataFrame(data_corr)
    dist_cosine = (1 / (1 +(pdist(df_corr, metric='cosine'))))
    pearsonCorrelationDict[name]=dist_cosine
```

```python
coseno = pd.DataFrame.from_dict(pearsonCorrelationDict, orient='index')
```

```python
coseno.columns = ["cero",'uno','dos','tres','cuatro','cinco','seis','siete','ocho','nueve']
```

```python
coseno_1=coseno.loc[:,['cero']]
```

```python
coseno_1.shape
```

```python
coseno_1.columns = ['usersimilares']
coseno_1['userId'] = coseno_1.index
coseno_1.index = range(len(coseno_1))
coseno_1.head()
```

```python
top_usuario=coseno_1.sort_values(by='usersimilares', ascending=False)[0:50]
top_usuario.head()
```

```python
top_rating_usuario=top_usuario.merge(ratings_df, left_on='userId', right_on='userId', how='inner')
top_rating_usuario.head()
```

```python
top_rating_usuario['pond_Rating'] = top_rating_usuario['usersimilares']*top_rating_usuario['rating']
top_rating_usuario.head()
```

```python
pond_top_rating_usuario = top_rating_usuario.groupby('movieId').sum()[['usersimilares','pond_Rating']]
pond_top_rating_usuario.columns = ['sum_usersimilares','sum_pond_Rating']
pond_top_rating_usuario.head()
```

```python
recommendation_df = pd.DataFrame()

recommendation_df['weighted average recommendation score'] = pond_top_rating_usuario['sum_pond_Rating']/pond_top_rating_usuario['sum_usersimilares']
recommendation_df['movieId'] = pond_top_rating_usuario.index
recommendation_df.head()
```

```python
recommendation_df = recommendation_df.sort_values(by='weighted average recommendation score', ascending=False)
recommendation_df.head(30)
print(movies_df.loc[movies_df['movieId'].isin(recommendation_df.head(30)['movieId'].tolist())])
```
